```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/world_demographics.csv.zip')
print(df.head())

```

```python
# Filter for total area and both sexes only
df_filtered = df[(df['Area'] == 'Total') & (df['Sex'] == 'Both Sexes')]

# Convert Age and Value columns to numeric
df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')
df_filtered['Value'] = pd.to_numeric(df_filtered['Value'], errors='coerce')

# Drop missing values
df_filtered = df_filtered.dropna(subset=['Age', 'Value'])

# Plot histogram: Age distribution weighted by population
plt.figure(figsize=(8, 4))
sns.histplot(data=df_filtered, x='Age', weights='Value', bins=20, color='grey', kde=False)
plt.title('Population Distribution by Age (World - All Genders)')
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.tight_layout()
plt.show()
```
